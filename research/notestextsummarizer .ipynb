### Basic Functionality Of Huggingface Model

```python
from transformers import AutoTokenizer, PegasusForConditionalGeneration
# AutoTokenizer: Used for automatically loading the appropriate tokenizer for the model.
# PegasusForConditionalGeneration: The specific Pegasus model architecture designed for text summarization.

model = PegasusForConditionalGeneration.from_pretrained("google/pegasus-xsum")
tokenizer = AutoTokenizer.from_pretrained("google/pegasus-xsum")

# from_pretrained("google/pegasus-xsum"): This loads a pre-trained Pegasus model and tokenizer specifically
# the "google/pegasus-xsum" variant, which is fine-tuned for extractive summarization.
ARTICLE_TO_SUMMARIZE = (
    "PG&E stated it scheduled the blackouts in response to forecasts for high winds "
    "amid dry conditions. The aim is to reduce the risk of wildfires. Nearly 800 thousand customers were "
    "scheduled to be affected by the shutoffs which were expected to last through at least midday tomorrow."
)
inputs = tokenizer(ARTICLE_TO_SUMMARIZE, max_length=1024, return_tensors="pt")

# tokenizer(...): The tokenizer processes the input text (ARTICLE_TO_SUMMARIZE) and converts it into numerical representations (tokens) that the model can understand.
# max_length=1024: Sets the maximum sequence length for the input. Longer sequences will be truncated.
# return_tensors="pt": Specifies that the output should be PyTorch tensors.

# Generate Summary
summary_ids = model.generate(inputs["input_ids"])

# model.generate(...): This line uses the loaded Pegasus model to generate the summary. It takes the tokenized input (inputs["input_ids"]) and produces the summary as token IDs (summary_ids).

tokenizer.batch_decode(summary_ids, skip_special_tokens=True, clean_up_tokenization_spaces=False)[0]

# tokenizer.batch_decode(...): This converts the generated token IDs (summary_ids) back into human-readable text.
# skip_special_tokens=True: Removes special tokens (like start and end tokens) from the output.
# clean_up_tokenization_spaces=False: Controls whether extra spaces introduced during tokenization should be removed.
# [0]: This extracts the first (and likely only) summary from the decoded output
```

### Preparing Data For Training For Sequence To Sequence Model

{

    'dialogue': "Hi! How are you?",
  
    'summary': "The speaker is asking how the
    other person is."
}

# The above data is need to convert into below three fields which is required in any sequence to sequence model to train

{

    'input_ids': [123, 456, 789, ...],  # Token IDs for the dialogue


    'attention_mask': [1, 1, 1, ...],  # Attention mask for the input
    #Attention mask is just to apply some special characters whthin this word in the form of token which is useful for sequence to sequence model to implement or predict
    
    'labels': [321, 654, 987, ...]  # Token IDs for the summary (target)
    #labels is nothing but it is a token ID for the summary target
}

```python
def convert_examples_to_features(example_batch):
    input_encodings = tokenizer(example_batch['dialogue'] , max_length = 1024, truncation = True )
# above input_encoding will give as input_ids and attention_mask
    with tokenizer.as_target_tokenizer():
        target_encodings = tokenizer(example_batch['summary'], max_length = 128, truncation = True )
# target encodings will give the label tokens
    return {
        'input_ids' : input_encodings['input_ids'],
        'attention_mask': input_encodings['attention_mask'],
        'labels': target_encodings['input_ids']
    }
```

```python
dataset_samsum_pt = dataset_samsum.map(convert_examples_to_features, batched = True)
# batched = True: This argument is passed to the map function. It indicates that the convert_examples_to_features function
# should be applied to batches of data examples at once instead of individually.
# This is often more efficient, especially when dealing with large datasets and using GPUs.
```

DataCollatorForSeq2Seq is a special data collator designed for sequence-to-sequence models (e.g., Pegasus, T5, BART) that helps in preparing batches of data for training.

```python
# Training

from transformers import DataCollatorForSeq2Seq

seq2seq_data_collator = DataCollatorForSeq2Seq(tokenizer, model=model_pegasus)
```

```python
from transformers import TrainingArguments, Trainer

trainer_args = TrainingArguments(
    output_dir='pegasus-samsum', num_train_epochs=1, warmup_steps=500,
    per_device_train_batch_size=1, per_device_eval_batch_size=1,
    weight_decay=0.01, logging_steps=10,
    evaluation_strategy='steps', eval_steps=500, save_steps=1e6,
    gradient_accumulation_steps=16
)
# here epocs size is 1 because of limited GPU power
"""trainer_args = TrainingArguments(...): This line creates an instance of the TrainingArguments class and stores it in the trainer_args variable. This object holds all the configurations for the training process.
Let's look at the individual arguments within TrainingArguments:

output_dir='pegasus-samsum': Specifies the directory where the trained model and checkpoints will be saved. In this case, it's 'pegasus-samsum'.
num_train_epochs=1: Sets the number of training epochs to 1. An epoch means one complete pass through the entire training dataset.
warmup_steps=500: Defines the number of warmup steps for the learning rate scheduler. During warmup, the learning rate is gradually increased from a small value to the initial learning rate.
per_device_train_batch_size=1: Sets the batch size for training on each device (GPU or CPU) to 1. Batch size is the number of training examples processed in one iteration.
per_device_eval_batch_size=1: Sets the batch size for evaluation on each device to 1.
weight_decay=0.01: Applies weight decay regularization with a factor of 0.01 to prevent overfitting.
logging_steps=10: Logs training progress every 10 steps.
evaluation_strategy='steps': Specifies that evaluation should be performed at regular intervals based on the number of training steps.
eval_steps=500: Evaluates the model every 500 training steps.
save_steps=1e6: Saves model checkpoints every 1 million training steps.
gradient_accumulation_steps=16: Accumulates gradients over 16 steps before updating the model's weights. This is a technique to simulate larger batch sizes when limited by GPU memory."""
```

```python
trainer = Trainer(model=model_pegasus, args=trainer_args,
                  tokenizer=tokenizer, data_collator=seq2seq_data_collator,
                  train_dataset=dataset_samsum_pt["test"],
                  eval_dataset=dataset_samsum_pt["validation"])
# here we have used test dataset instead of train because test dataset is small
```

```python
trainer.train()
```

```python
# Evaluation
### lst[1,2,3,4,5,6]-> [1,2,3][4,5,6]
def generate_batch_sized_chunks(list_of_elements, batch_size):
    """split the dataset into smaller batches that we can process simultaneously
    Yield successive batch-sized chunks from list_of_elements."""
    for i in range(0, len(list_of_elements), batch_size):
        yield list_of_elements[i : i + batch_size]



def calculate_metric_on_test_ds(dataset, metric, model, tokenizer,
                               batch_size=16, device=device,
                               column_text="article",
                               column_summary="highlights"):
    article_batches = list(generate_batch_sized_chunks(dataset[column_text], batch_size))
    target_batches = list(generate_batch_sized_chunks(dataset[column_summary], batch_size))

    for article_batch, target_batch in tqdm(
        zip(article_batches, target_batches), total=len(article_batches)):

        inputs = tokenizer(article_batch, max_length=1024,  truncation=True,
                        padding="max_length", return_tensors="pt")

        summaries = model.generate(input_ids=inputs["input_ids"].to(device),
                         attention_mask=inputs["attention_mask"].to(device),
                         length_penalty=0.8, num_beams=8, max_length=128)
        ''' parameter for length penalty ensures that the model does not generate sequences that are too long. '''

        # Finally, we decode the generated texts,
        # replace the  token, and add the decoded texts with the references to the metric.
        decoded_summaries = [tokenizer.decode(s, skip_special_tokens=True,
                                clean_up_tokenization_spaces=True)
               for s in summaries]

        decoded_summaries = [d.replace("", " ") for d in decoded_summaries]


        metric.add_batch(predictions=decoded_summaries, references=target_batch)

    #  Finally compute and return the ROUGE scores.
    score = metric.compute()
    return score
```

### Interpreting Good vs. Bad ROUGE Scores:
1. Scores close to 1: This indicates a strong overlap between the generated summary and the reference summary, which is desirable in summarization tasks. For example, an F1-score of 0.7 or higher across metrics is generally considered good.
2. Scores between 0.5 and 0.7: Indicates moderate overlap. The summary might be capturing key points but is likely missing some structure or important information.
3. Scores below 0.5: Suggest a poor match between the generated and reference summaries. The model might be generating irrelevant or incomplete summaries that don’t capture the key ideas well.
